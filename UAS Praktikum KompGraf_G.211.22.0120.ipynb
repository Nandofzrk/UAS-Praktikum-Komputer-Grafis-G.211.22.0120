```python
import pandas as pd

# Membuat DataFrame Dividen/Share
data = {
    'Stock Name': ['Stock A', 'Stock B', 'Stock C', 'Stock D', 'Stock E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Dividen/Share': [0.5, 0.8, 0.3, 0.9, 0.6]
}

df = pd.DataFrame(data)

# Mengurutkan DataFrame berdasarkan Dividen/Share secara menurun
df_sorted = df.sort_values(by='Dividen/Share', ascending=False)

# Mengambil 5 entitas teratas
top_5_dividends = df_sorted.head(5)

# Menampilkan hasil
print("Top 5 Stocks with Highest Dividen/Share:")
print(top_5_dividends[['Stock Name', 'Symbol']])
```

```python
import pandas as pd

# Membuat DataFrame yield
data = {
    'Stock Name': ['Stock A', 'Stock B', 'Stock C', 'Stock D', 'Stock E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Annual Yield': [0.05, 0.08, 0.03, 0.09, 0.06]
}

df = pd.DataFrame(data)

# Mengurutkan DataFrame berdasarkan Annual Yield secara menurun
df_sorted = df.sort_values(by='Annual Yield', ascending=False)

# Mengambil 7 entitas teratas
top_7_annual_yield = df_sorted.head(7)

# Menampilkan hasil
print("Top 7 Stocks with Highest Annual Yield:")
print(top_7_annual_yield[['Stock Name', 'Symbol']])
```
